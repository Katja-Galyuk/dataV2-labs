# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
comments = SITE.fetch('comments')
comments
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
questions = SITE.fetch('questions', fromdate=1580515200, todate=1583020799)
answers = SITE.fetch('answers', fromdate=1580515200, todate=1583020799)
questions
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
voted_question = SITE.fetch('questions', fromdate=1583712000, todate=1583798399, min=5, tagged='python', sort='votes')['items']
voted_question[0]

```
